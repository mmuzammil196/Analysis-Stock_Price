## Question 6 
Plot in the same figure the boxplot of High prices for the whole period of all four stocks.  

```python
data = [Apple['High'], Microsoft['High'], Google['High'], Amazon['High']]
labels = ['Apple', 'Microsoft', 'Google', 'Amazon']
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(15,7))
box = plt.boxplot(data, patch_artist=True, showfliers=False)
 
colors = ['red', 'blue', 'yellow', 'orange']
 
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)
plt.xticks([1,2,3,4], labels)
plt.ylabel('Stock Price')
plt.title('High Stock Prices of Apple, Microsoft, Google and Amazon')
plt.legend(labels)
plt.show()
```
